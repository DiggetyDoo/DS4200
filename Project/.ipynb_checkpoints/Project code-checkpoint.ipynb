```python
import pandas as pd

# List of CSV files
files = ["female_coaches.csv","female_players.csv", "female_teams.csv", "male_coaches.csv", 
         "male_players.csv", "male_teams.csv"]

# Dictionary to store dataframes
dataframes = {}

# Read each CSV file and create a dataframe
for file in files:
    # Generate dataframe name by removing the extension
    dataframe_name = file.split(".")[0]
    
    # Read CSV file into dataframe
    dataframes[dataframe_name] = pd.read_csv(file)

# Access dataframes using keys
female_coaches_df = dataframes["female_coaches"]
female_players_df = dataframes["female_players"]
female_teams_df = dataframes["female_teams"]
male_coaches_df = dataframes["male_coaches"]
male_players_df = dataframes["male_players"]
male_teams_df = dataframes["male_teams"]
```

```python
male_players_df
```

```python
#Function to delete unwanted columns & clean the position ratings

def clean_player(dataframe):
    
    del_col = ['player_url', 'fifa_update', 'update_as_of',
               'dob', 'club_jersey_number', 'club_loaned_from',
               'club_joined_date', 'club_contract_valid_until_year', 'nation_team_id',
               'nation_position', 'nation_jersey_number', 'preferred_foot', 'real_face']
    
    cleaned_dataframe = dataframe.drop(columns = del_col)
    
    target_col = ['ls', 'st', 'rs', 'lw', 'lf', 'cf',
                  'rf', 'rw', 'lam', 'cam', 'ram', 'lm',
                  'lcm', 'cm', 'rcm', 'rm', 'lwb', 'ldm',
                  'cdm', 'rdm', 'rwb', 'lb', 'lcb', 'cb',
                  'rcb', 'rb', 'gk']
    
    for column in target_col:
        cleaned_values = []
        for value in cleaned_dataframe[column]:
            if '+' in str(value):
                parts = value.split('+')
                cleaned_values.append(int(parts[0]) + int(parts[1]))
                
            elif '-' in str(value):
                parts = value.split('-')
                cleaned_values.append(int(parts[0]) - int(parts[1]))
            
            else:
                cleaned_values.append(value)
        cleaned_dataframe[column] = cleaned_values

    return cleaned_dataframe

   

    
```

```python
clean_female_players_df = clean_player(female_players_df)
clean_female_players_df.head()
```

```python
clean_male_players_df = clean_player(male_players_df)
clean_male_players_df.tail()
```

```python
#Selecting fifa version and league level

def version_league(dataframe, selected_fifa_version, selected_league_level):
    
    filtered_dataframe = dataframe[(dataframe['fifa_version'] == selected_fifa_version) & (dataframe['league_level'] == selected_league_level)]
    
    return filtered_dataframe
```

```python
male_player_24_1 = version_league(clean_male_players_df, 24.0, 1)
male_player_24_1
```

```python
#Selecting players by position
def pos_filter(dataframe, selected_position):
   
    filtered_dataframe = dataframe[dataframe['player_positions'].str.contains(selected_position)]
    return filtered_dataframe
```

```python
male_player_24_1_cm = pos_filter(male_player_24_1, 'CM')
male_player_24_1_cm
```
