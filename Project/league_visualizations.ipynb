```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV file
file_path = 'male_players.csv'
data = pd.read_csv(file_path)

# Extracting primary position and mapping broader position categories
data['primary_position'] = data['player_positions'].apply(lambda x: x.split(",")[0])
position_mapping = {
    'ST': 'Forward', 'LW': 'Forward', 'RW': 'Forward', 'CF': 'Forward',
    'CAM': 'Midfielder', 'CM': 'Midfielder', 'CDM': 'Midfielder', 'RM': 'Midfielder', 'LM': 'Midfielder',
    'CB': 'Defender', 'LB': 'Defender', 'RB': 'Defender', 'LWB': 'Defender', 'RWB': 'Defender',
    'GK': 'Goalkeeper'
}
data['position_category'] = data['primary_position'].map(position_mapping)
data = data[data["league_name"].isin(["Premier League", "Bundesliga", "La Liga"])]
sns.set_style("whitegrid")
```

```python

# Calculating average player age and overall rating for each league
league_averages = data.groupby('league_name').agg({'age': 'mean', 'overall': 'mean'}).reset_index()

# Sorting the leagues by average age for visualization
league_averages_sorted_by_age = league_averages.sort_values(by='age')

# Plotting Average Player Age per League
plt.figure(figsize=(10, 8))
sns.barplot(x='age', y='league_name', data=league_averages_sorted_by_age, palette="coolwarm")
plt.title('Average Player Age per League')
plt.xlabel('Average Age')
plt.ylabel('League')
plt.show()

# Sorting the leagues by average overall rating for visualization
league_averages_sorted_by_rating = league_averages.sort_values(by='overall', ascending=False)

# Plotting Average Overall Rating per League
plt.figure(figsize=(10, 8))
sns.barplot(x='overall', y='league_name', data=league_averages_sorted_by_rating, palette="coolwarm")
plt.title('Average Overall Rating per League')
plt.xlabel('Average Overall Rating')
plt.ylabel('League')
plt.show()
```

```python

# Creating box plots for the distribution of player ages across leagues
plt.figure(figsize=(15, 10))
sns.boxplot(x='age', y='league_name', data=data, palette="coolwarm")
plt.title('Distribution of Player Ages per League')
plt.xlabel('Age')
plt.ylabel('League')
plt.show()
```

```python
import ipywidgets as widgets
from IPython.display import display
import pandas as pd
import altair as alt

alt.data_transformers.enable('json')
# Assuming 'data' is your DataFrame with 'age' and 'overall' columns

# Function to update chart based on slider
def update_chart(overall_rating):
    filtered_data = data[data['overall'] > overall_rating]
    chart = alt.Chart(filtered_data).mark_bar().encode(
        x=alt.X('age', bin=alt.Bin(maxbins=27)),
        y='count()',
    ).properties(
        width=600,
        height=400
    )
    display(chart)

# Create slider widget
slider = widgets.IntSlider(
    value=data['overall'].min(),
    min=data['overall'].min(),
    max=data['overall'].max(),
    step=1,
    description='Min Overall:',
    continuous_update=False
)

# Display slider and chart
widgets.interactive(update_chart, overall_rating=slider)
```
